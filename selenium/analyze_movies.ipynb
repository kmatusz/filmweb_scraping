```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('movies.csv', dtype='object')
a.head()
```

```python
a = a.drop(['Unnamed: 0'], axis = 1)
```

```python
a.head()
```

```python
a.isna().sum()
```

```python
a['movie_rating_avg'] = a.movie_rating_avg.apply(lambda x: float(x.replace(',', '.').replace(' ', '')))
a['movie_rating_no'] = a.movie_rating_no.apply(lambda x: float(x.replace(',', '.').replace(' ', '')))

a['user_rating'] = a.user_rating.astype(np.float)
```

```python
a.head()
```

```python
months_mapping = {
    'sty': '1',
    'lut': '2',
    'mar': '3',
    'kwi': '4',
    'maj': '5',
    'cze': '6',
    'lip': '7',
    'sie': '8',
    'wrz': '9',
    'paź': '10',
    'lis': '11',
    'gru': '12'
}
```

```python
current_year = '2020'
```

```python
temp_dates = pd.DataFrame({'full': a.user_rated_date.apply(lambda x: x.split(' ')).apply(lambda x: x + [current_year] if len(x)==2 else x )})#.apply(lambda x: x[1]).value_counts()
```

```python
temp_dates['Day'] = temp_dates.full.apply(lambda x: x[0])
temp_dates['Year'] = temp_dates.full.apply(lambda x: x[2])
temp_dates['Month'] = temp_dates['full'].apply(lambda x: x[1][:3]).map(months_mapping)
```

```python
temp_dates.head()
```

```python
a.user_rated_date = pd.to_datetime(temp_dates.drop(['full'], axis = 1))
```
